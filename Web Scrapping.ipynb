```python
import requests 
from bs4 import BeautifulSoup 
```

```python
URL = "https://summerofcode.withgoogle.com/archive/2018/organizations/"
r = requests.get(URL) 
```

```python
soup = BeautifulSoup(r.content, 'html5lib') 
print(soup.prettify()) 
```

```python
whatweneed = soup.findAll('a',class_= "organization-card__link")
print(whatweneed)
```
